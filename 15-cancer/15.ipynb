```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

# USE NP.NAN

```python
df=pd.read_csv('BreastCancerWc.csv')
df
```

```python
cols=['Sample code number','Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class']
```

```python
df.columns=cols
```

```python
df
```

# 1

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.replace('NAType', np.nan, inplace=True)
df=df.astype(float)
df.dropna(inplace=True)
```

```python
df
```

# 2 

```python
Z_val=np.abs((df-df.mean())/df.std())
out=Z_val > 3
df=df[~out.any(axis=1)]
df
```

# 3

```python
# Scaler=StandardScaler()
# df=Scaler.fit_transform(df)
# df
```

```python
df.loc[df['Class']==2,'Class']='Y'
df.loc[df['Class']==4,'Class']='N'
df
```

```python
lb=LabelEncoder()
```

```python
df['Class']=lb.fit_transform(df['Class'])
df
```

# 4

```python
X = df.drop('Class', axis=1)  # Features
Y = df['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=46)
```

```python
reg=LogisticRegression()
reg.fit(X_train,Y_train)
```

```python
a1=reg.predict(X_test)
```

```python
nvb=MultinomialNB()
nvb.fit(X_train,Y_train)
```

```python
nvbg=GaussianNB()
nvbg.fit(X_train,Y_train)
```

```python
a2=nvb.predict(X_test)
a3=nvbg.predict(X_test)
```

```python
z1=accuracy_score(Y_test,a1)
z2=accuracy_score(Y_test,a2)
z3=accuracy_score(Y_test,a3)
```

```python
print(z1)
print(z2)
print(z3)
```
