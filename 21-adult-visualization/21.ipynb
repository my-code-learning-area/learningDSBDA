```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sn
```

```python
df = pd.read_csv("adult.csv")
df
```

# HIST

```python
df.hist(['age','education-num','hours-per-week'], bins=5, figsize=(20, 15))
mat.tight_layout()
```

# PIE

```python
df['salary'].value_counts().plot(kind='pie', labels=['male', 'female'], autopct="%.2f%%")
```

# DOT

```python
sn.stripplot(data=df, x='workclass', y='education', jitter=True)
```

# BAR

```python
df['salary'].value_counts().plot(kind="bar", figsize=(6,4))
```

# Line

```python
mat.plot(df['age'])
```

# Scatterplot

```python
sn.scatterplot(data=df, x='workclass', y='fnlwgt', hue="workclass")
```
