```python
import requests ## used to send http request to specified url
import bs4 ## pull the data from web
```

```python
request1 = requests.get("https://www.flipkart.com/poco-c31-royal-blue-32-gb/p/itm9656ebd21e0b3?pid=MOBG73E7ZBHHQ3RB&lid=LSTMOBG73E7ZBHHQ3RBQ7POR7&marketplace=FLIPKART&q=poco+c31+royal+blue&store=tyy%2F4io&srno=s_1_1&otracker=search&otracker1=search&fm=organic&iid=520aea99-8bc8-4283-ade2-0fd5317ca669.MOBG73E7ZBHHQ3RB.SEARCH&ppt=None&ppn=None&ssid=3kvv6grsww0000001685263969710&qH=ae2909a05b885afa")
request1
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.findAll("div", {'class':'t-ZTKy'})
for review in reviews:
    print(review.get_text()+"\n\n")
```

```python
## average / overall rating
ratings = soup.find("div", {"class":"_3LWZlK"})
ratings.getText()
```

```python
individualRatings = soup.findAll("div", {"class":"_3LWZlK _1BLPMq"})
for rating in individualRatings:
    print(rating.getText()+"\n")
```

```python
tags = soup.find("span", {"class":"yhB1nd GXgmTe"})
tags.getText()
```

```python
customers = soup.findAll("p", {"class":"_2sc7ZR _2V5EHH"})
for customer in customers:
    print(customer.getText() + "\n")
```

```python
offers = soup.findAll("li", {"class":"_16eBzU col"})
for offer in offers:
    print(offer.getText() + "\n\n")
```
