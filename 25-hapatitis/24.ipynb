```python
import pandas as pd
```

```python
DataFrame=pd.read_csv('hepatitis.csv')
DataFrame.columns = ["Class","AGE","SEX","STEROID","ANTIVIRALS","FATIGUE","MALAISE","ANOREXIA","LIVER BIG","LIVER FIRM","SPLEEN PALPABLE","SPIDERS","ASCITES","VARICES","BILIRUBIN","ALK PHOSPHATE","SGOT","ALBUMIN","PROTIME","HISTOLOGY"]
DataFrame.head()
```

```python
DataFrame['SEX'].unique()
```

```python
SEX1=DataFrame[DataFrame['SEX']==1]
SEX1.head()
```

```python
SEX2=DataFrame[DataFrame['SEX']==2]
SEX2.head()
```

# Merge

```python
Merged1=pd.concat([SEX1,SEX2])
Merged1.head()
```
