```python
import pandas as pd
import numpy as np
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

# Remove NULL and Negative values

```python
df = df.dropna()
df.head()
```

```python
df = df.replace("?", pd.NA)
df.head()
```

```python
df = df[df >= 0]
df.head()
```

# Outlier Detection and Remove

## Detection

```python
df.std()
```

```python
df.mean()
```

```python
out = np.abs((df - df.mean()) / df.std())
out
```

```python
lier = out > 3
lier.sum()
```

## removing

```python
# df[~lier.any(axis=1)]
```

```python
df = df[~lier]
df
```

# Transformation

```python
df.loc[df['sex']==1,'sex']='M'
df.loc[df['sex']==0,'sex']='F'
df
```

```python
lbl = LabelEncoder()
lbl
```

```python
df['sex'] = lbl.fit_transform(df['sex'])
df
```

# Build data model using regression and knn

**creating input and result**

```python
x=df.drop('target', axis=1)
x
```

```python
y=df['target']
y
```

**create train and test datasets**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=5)
```

## LogisticRegression

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
r1 = lr.predict(x_test)
r1
```

```python
r1Acc = accuracy_score(y_test, r1)
r1Acc
```

# KNN

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
r2 = knn.predict(x_test)
r2
```

```python
r2Acc = accuracy_score(y_test, r2)
```
