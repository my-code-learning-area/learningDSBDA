```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import seaborn as sb
```

```python
df = pd.read_csv("heart.csv")
df
```

```python
df.isnull().sum()
```

# HISTOGRAM

```python
cols = ["age", "sex", "oldpeak"]
df[cols].hist(bins=5,figsize=(20, 15))
plt.tight_layout() #optional
```

# PIE CHART

```python
df['sex'].value_counts().plot(kind="pie", labels=['male', 'female'], autopct="%1.2f%%")
plt.tight_layout() # optional
```

# BOX

```python
sb.boxplot(data=df, x='target', y='trestbps')
```

# Scatter Plot

```python
sb.scatterplot(data=df, x='chol', y='thalach', hue="target")
```

```python
df['fbs'].value_counts().plot(kind="bar")
plt.tight_layout() #optional
```
