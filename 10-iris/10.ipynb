```python
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

# Subset

```python
df['class'].unique()
```

```python
irissetosa = df[df['class']=='Iris-setosa']
irissetosa.head()
```

```python
irisversicolor = df[df['class'] == 'Iris-versicolor']
irisversicolor.head()
```

```python
irisvirginica = df[df['class'] == 'Iris-virginica']
irisvirginica.head()
```

# merge

```python
merged = pd.concat([irissetosa,irisversicolor])
merged.head()
```

# sort

```python
sortedData = df.sort_values('petal length in cm')
sortedData.head()
```

# transpose

```python
df.T
```

# melting

```python
melt = df.melt(id_vars=['sepal length in cm', 'petal length in cm'], var_name='Metrics', value_name='otherValues')
melt.head()
```

# casting

```python
cast = pd.pivot_table(df,index=['sepal length in cm', 'petal length in cm'], values=['sepal width in cm', 'petal width in cm'])
cast.head()
```
