```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sn
```

```python
df = pd.read_csv("heart.csv")
df
```

```python
df.isna().sum()
```

# HIST - 1

```python
cols = ['age', 'chol', 'thalach']
df[cols].hist(bins=15, figsize=(15,10))
mat.tight_layout()
```

# DOT - 0

```python
mat.figure(figsize=(10,6))
sn.stripplot(data=df, x="cp", y="trestbps", jitter=True)
mat.xlabel('cp')
mat.ylabel('trestbps')
mat.plot()
```

# BAR - 1

```python
col = 'restecg'
df[col].value_counts().plot(kind="bar")
mat.xlabel(col)
mat.ylabel('Count')
mat.show() ## ?""
```

```python
col = 'fbs'
df[col].value_counts().plot(kind="bar")
mat.xlabel(col)
mat.ylabel("Count")
```

```python
col = 'slope'
df[col].value_counts().plot(kind="bar")
mat.xlabel(col)
mat.ylabel("Count")
```

# Line

```python
mat.plot(df['trestbps'], color='#3299ff')
```

```python
col = 'oldpeak'
mat.plot(df[col])
```

```python
col = 'thalach'
mat.plot(df[col])
```

# Multivariate analysis using scatter plot

```python
# for 2 elected cols
sn.scatterplot(data=df, x='chol', y='thalach', hue='target')
```

# PIE

```python
df['sex'].value_counts().plot(kind="pie", labels=["male", "female"], autopct='%1.2f%%')
```

# BOX

```python
sn.boxplot(data=df, x='age', y='oldpeak')
```
